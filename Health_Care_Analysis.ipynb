# Data Cleaning

```python
# Lets check if there are any missing values within the data
print('Missing Values:')
df.isnull().sum()

# For this data no missing value found but if found handle them appropriately.
```

```python
# Check for any duplicates with in data
print('Duplicate Rows:')
df.duplicated().sum()
```

```python
# Look at dduplicated rows within data frame
df[df.duplicated()]
```

```python
# Remove duplicated rows as they will affect analysis
df.drop_duplicates(inplace=True)
```

```python
# Fix capitalization of patient names within the Name column
df["Name"] = df["Name"].str.title()
```

```python
# Replace spacing within the column with an underscore
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Check data turnout after cleaning
df.head()
```

```python
# Convert columns containig dates to datetime
df['Date_of_Admission'] = pd.to_datetime(df['Date_of_Admission'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])
```

```python
# Checking if columns were converted
df.info()
```

# Data Analysis

```python
# Statistical Summary of Age
print('Age Summary:')
df.Age.describe()
```

```python
# Average age of patients
print('Average Age:')
df.Age.mean()
```

```python
# Most common blood type among patients
print('Most Common Blood Type:')
df["Blood_Type"].value_counts().sort_values(ascending=False)
```

```python
# Distribution of Gender among patients
print('Gender Distribution:')
df.Gender.value_counts()
```

```python
# Summary of age statistics for each medical condition
print('Medical Condition Statistics:')
df.groupby("Medical_Condition")['Age'].describe()
```

```python
# Hospitals with the most admissions
print('Hospital with Most Admissions:')
df.Hospital.value_counts().sort_values(ascending=False).head(5)
```

```python
# Number of patients admitted per admission type
print('Admission Type Distribution:')
df.Admission_Type.value_counts()
```

```python
# Docotors with the most visits
print('Doctors with Most Visits:')
df.Doctor.value_counts().head(5)
```

Average stay of patient within a hospital

```python
# Avergae stay of patient within assigned hospital
print('Average Stay per Hospital:')
(df["Discharge_Date"] - df["Date_of_Admission"]).mean()
```

```python
# Average stay of patients within a hospital based on their medical conditions
print('Average Stay per Medical Condition:')
df["average_discharge"] = (df["Discharge_Date"] - df["Date_of_Admission"]).dt.days
df.groupby("Medical_Condition")["average_discharge"].mean()
```

```python
# Peak Months for each year of admitted patients
print('Peak Months for Each Year:')
df["month"] = df["Date_of_Admission"].dt.month_name()
df["year"] = df["Date_of_Admission"].dt.year
month_counts = df.groupby(["year", "month"]).size().reset_index(name="Admissions")
peak_month = month_counts.sort_values(['year', 'Admissions'], ascending = [True, False]).drop_duplicates('year')
print(peak_month)
```

```python
# Total admitted patients for each year
print('Total Admitted Patients per Year:')
df.year.value_counts()
```

```python
# Total number of times each medication was given for each test results
print('Medication Usage per Test Result:')
pd.crosstab(df["Medication"], df["Test_Results"])
```
